```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")

print("Path to dataset files:", path)
```

```python
import pandas as pd
books_data = pd.read_csv(f"{path}/books.csv")
books_data
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ax = plt.axes()

sns.heatmap(books_data.isna().transpose(), cbar=False, ax=ax)

plt.xlabel("Missing Values")
plt.ylabel("Columns")

plt.show()
```

### As the above chart shows that there might be correlation between average_ratings, num_pages or ratings_count we want to see that if that correlation exist with the description or not. Because otherwise it will effect the training of model.

```python
import numpy as np
books_data["missing_description"] = np.where(books_data["description"].isna() , 1 , 0)
books_data["age_of_book"] = 2025 - books_data["published_year"]
```

```python
columns_of_interest = ["missing_description", "age_of_book", "num_pages", "average_rating"]

correlation_matrix = books_data[columns_of_interest].corr(method = "spearman") # We are using spearman as it is more appopriate for non continous values which in our case is binary for missing_description

sns.set_theme(style="white")
plt.figure(figsize=(8, 6))
heatmap = sns.heatmap(correlation_matrix, annot=True)

heatmap.set_title("Correlation heatmap")

plt.show()
```

### As the above figure shows that the correlation between different columns is not strong (it is strong if it close to 1 or -1). So we don't have to care about other columns.

```python
# Data where none of the following column is missing collectively
book_missing = books_data[~(books_data["description"].isna()) &
    ~(books_data["num_pages"].isna()) &
    ~(books_data["average_rating"].isna()) &
    ~(books_data["published_year"].isna())
]

book_missing.head()
```

### As we will be using description in our recommendation system. Here we will be checking the lengths of description and selecting only those which are meaningful.

```python
book_missing["words_in_description"] = books_data["description"].str.split().str.len() # Splitting every word when encountering space
```

```python
# NOT Useful as the description with 20 words are not enough
book_missing.loc[ book_missing["words_in_description"].between(1,20) , "description" ]
```

```python
book_missing_25_words_or_more = book_missing[book_missing["words_in_description"] >= 25]   # More than 25 words description
```

### Now we will be cleaning the subtitle field as there are lot of Nan values there.

```python
book_missing_25_words_or_more["title_subtitle"] = np.where(book_missing_25_words_or_more["subtitle"].isna(), 
                                                            book_missing_25_words_or_more["title"],
                                                            book_missing_25_words_or_more["title"] + ": " + book_missing_25_words_or_more["subtitle"]
                                                            )
```

```python
book_missing_25_words_or_more["tagged_description"] = book_missing_25_words_or_more[["isbn13", "description"]].astype(str).agg(" ".join, axis=1)
book_missing_25_words_or_more
```

```python
# Drop unnecessary columns and storing results in csv
book_missing_25_words_or_more.drop(["subtitle", "missing_description", "age_of_book", "words_in_description"], axis=1).to_csv("books_cleaned.csv", index = False)
```
